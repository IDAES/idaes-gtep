```python
import pandas as pd
```

```python
gen_nonsense = pd.read_csv("Bus_data_gen_weights_mappings.csv")
gen_nonsense.head()
```

```python
gen_nonsense = gen_nonsense[gen_nonsense["Gen bus/ Non-gen bus"] == 1]
gen_nonsense.head()
```

```python
candidate_ng_ct_df = gen_nonsense[gen_nonsense["Natural Gas_CT - Generation Weight"] > 100]
len(candidate_ng_ct_df["Bus Number"])
```

```python
candidate_ng_fe_df = gen_nonsense[gen_nonsense["Natural Gas_FE - Generation Weight"] > 300]
len(candidate_ng_fe_df["Bus Number"])
```

```python
candidate_pv_df = gen_nonsense[gen_nonsense["Solar - Utility PV - Generation Weight"]> 180]
len(candidate_pv_df["Bus Number"])
```

```python
candidate_wind_df = gen_nonsense[gen_nonsense["Land-Based Wind - Generation Weight"]> 150]
len(candidate_wind_df["Bus Number"])
```

```python
candidate_ng_ct_by_bus = dict(zip(candidate_ng_ct_df["Bus Name"], candidate_ng_ct_df["Bus Number"]))
candidate_ng_fe_by_bus = dict(zip(candidate_ng_fe_df["Bus Name"],candidate_ng_fe_df["Bus Number"]))
candidate_pv_by_bus = dict(zip(candidate_pv_df["Bus Name"], candidate_pv_df["Bus Number"]))
candidate_wind_by_bus = dict(zip(candidate_wind_df["Bus Name"], candidate_wind_df["Bus Number"]))
```

```python
target_columns = ["GEN UID","Bus ID","Unit Type","Fuel","MW Inj","MVAR Inj","V Setpoint p.u.","PMax MW","PMin MW","QMax MVAR","QMin MVAR","Min Down Time Hr","Min Up Time Hr","Ramp Rate MW/Min","Start Time Cold Hr","Start Time Warm Hr","Start Time Hot Hr","Start Heat Cold MBTU","Start Heat Warm MBTU","Start Heat Hot MBTU","Non Fuel Start Cost $","Fuel Price $/MMBTU","Output_pct_0","Output_pct_1","Output_pct_2","Output_pct_3","Output_pct_4","HR_avg_0","HR_incr_1","HR_incr_2","HR_incr_3","HR_incr_4"]
unit_types_map = {"Nuclear":"NUC", "Natural Gas": "CT", "Coal": "COAL", "Wind": "WIND", "Solar": "PV", "Hydro": "HYDRO"}
fuel_types_map = {"NUC": "N", "COAL": "C", "CT": "G", "WIND": "W", "HYDRO": "H", "PV": "S"}
up_time_map = {
    "G": 6,
    "W": 0,
    "C": 24,
    "H": 0,
    "N": 48,
    "S": 0,
}
down_time_map = {
    "G": 8,
    "W": 0,
    "C": 12,
    "H": 0,
    "S": 0,
    "N": 48,
}
gen_data_target = pd.DataFrame(columns=target_columns)
gen_data_target["GEN UID"] = gen_uids
gen_data_target["Bus ID"] = bus_uids
gen_data_target["Unit Type"] = unit_types
gen_data_target["Fuel"] = fuels
gen_data_target["PMax MW"] = pmax
gen_data_target["PMin MW"] = pmin
gen_data_target["V Setpoint p.u."].fillna(1, inplace=True)
gen_data_target["Min Up Time Hr"] = minup
gen_data_target["Min Down Time Hr"] = mindown
gen_data_target["Output_pct_0"].fillna(0.6, inplace=True)
gen_data_target["Output_pct_1"].fillna(1,inplace=True)
gen_data_target["capex1"] = capex1
gen_data_target["capex2"] = capex2
gen_data_target["capex3"] = capex3
gen_data_target["fuel_cost1"] =fuel_costs1
gen_data_target["fuel_cost2"] =fuel_costs2
gen_data_target["fuel_cost3"] =fuel_costs3
gen_data_target["fixed_ops1"] = fixed_ops1
gen_data_target["fixed_ops2"] = fixed_ops2
gen_data_target["fixed_ops3"] = fixed_ops3
gen_data_target["var_ops1"] = var_ops1
gen_data_target["var_ops2"] = var_ops2
gen_data_target["var_ops3"] = var_ops3
gen_data_target.fillna("", inplace=True)
```
